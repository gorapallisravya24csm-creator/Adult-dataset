```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import StandardScaler
```

```python
print("Loading Adult Dataset...")

df = pd.read_csv(r"C:\Users\anits-csm\Downloads\adult\adult.data")

print(df.head())
print("Shape:", df.shape)
```

```python
df.replace('?', np.nan, inplace=True)

print(df.isnull().sum())
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

categorical_cols = df.select_dtypes(exclude=[np.number]).columns

print("Numeric Columns")
print(numeric_cols)

print("\nCategorical Columns")
print(categorical_cols)
```

```python
# Numeric Columns
num_imputer = SimpleImputer(strategy='mean')
df[numeric_cols] = num_imputer.fit_transform(df[numeric_cols])

# Categorical Columns
cat_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])

print("Missing Values Handled")
```

```python
z_scores = np.abs(zscore(df[numeric_cols]))

df = df[(z_scores < 3).all(axis=1)]

print("Shape After Removing Outliers")
print(df.shape)
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)

df.drop_duplicates(inplace=True)

print("Duplicates Removed")
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print(df.head())
```

```python
scaler = StandardScaler()

df[df.columns] = scaler.fit_transform(df)

print(df.head())
```

```python
scaler = StandardScaler()

df[df.columns] = scaler.fit_transform(df)

print(df.head())
```

```python
selector = VarianceThreshold(threshold=0.01)

reduced = selector.fit_transform(df)

selected_columns = df.columns[selector.get_support()]

df_reduced = pd.DataFrame(reduced, columns=selected_columns)

print("Reduced Shape")
print(df_reduced.shape)
```

```python
df_reduced.to_csv("adult_preprocessed.csv", index=False)

print("Adult Dataset Saved Successfully")
```
